```python
import supervision as sv
from ultralytics import YOLO
import numpy as np
```

```python
model = YOLO(r"..\models\yolov10n.pt")
tracker = sv.ByteTrack()
frames_generator = sv.get_video_frames_generator(r"..\videos\vehicle-counting-low.mp4")
start, end = sv.Point(x=0, y=1080), sv.Point(x=3840, y=1080)
line_zone = sv.LineZone(start=start, end=end)

for frame in frames_generator:
    result = model(frame)[0]
    detections = sv.Detections.from_ultralytics(result)
    detections = tracker.update_with_detections(detections)
    crossed_in, crossed_out = line_zone.trigger(detections)
    if np.any(crossed_in) or np.any(crossed_out):
        print("someone crossed")
```
